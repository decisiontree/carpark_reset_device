```python
def formatFig(ax,df,windowstatus=0):
    ax.plot(df.time_local, df.x, color='skyblue', linewidth=3,  label='x' )
    ax.plot(df.time_local, df.y, color='olive', linewidth=3, label='y' )
    ax.plot(df.time_local, df.z, color='red', linewidth=3, label='z' )
    if(windowstatus==1):
        ax.plot(df.time_local, df.binary_signal_status, 'o', color='green', linewidth=4, label='binary signal status' )

    ax.legend(loc='upper left')
    ax.tick_params(axis='x', rotation=70)
    if (df.dev_id.nunique() == 1) & (df.action.nunique() == 1):
        device = df.iloc[0].dev_id
        activity = df.iloc[0].action
        experimentID = df.iloc[0].experimentID
    else:
        print("Error, there are multiple device or activity values")
    
    ax.set_title('Experiment:{}, Device:{}, activity:{}'.format(experimentID, device,activity))
    ax.xaxis.set_major_formatter(DateFormatter('%H:%M:%S', tz=tz))
    ax.xaxis.set_major_locator(plt.MaxNLocator(df.shape[0]))

    #plt.savefig('{}.pdf'.format(experimentID))

def formatFigTarget(ax,df):
    ax.plot(df.time_local, df.x, color='skyblue', linewidth=3,  label='x' )
    ax.plot(df.time_local, df.y, color='olive', linewidth=3, label='y' )
    ax.plot(df.time_local, df.z, color='red', linewidth=3, label='z' )
    
    ax.legend(loc='upper left')
    ax.tick_params(axis='x', rotation=70)
    if (df.dev_id.nunique() == 1) & (df.action.nunique() == 1):
        device = df.iloc[0].dev_id
        activity = df.iloc[0].action
        experimentID = df.iloc[0].experimentID
    else:
        print("Error, there are multiple device or activity values")
    
    ax.set_title('Device:{}, activity:{}'.format(device,activity))
    ax.xaxis.set_major_formatter(DateFormatter('%H:%M:%S', tz=tz))
    ax.xaxis.set_major_locator(plt.MaxNLocator(df.shape[0]))

    
    
def calShift(df):
    df['xshift'] = (df.x - df.x.shift())  
    df['yshift'] = (df.y - df.y.shift())  
    df['zshift'] = (df.z - df.z.shift())
        
    df['time_localshift'] = (df.time_local - df.time_local.shift()).apply(lambda x:x.total_seconds())
    df['xyz_absoluteshift'] = df['xshift'].abs() + df['yshift'].abs() + df['zshift'].abs()
    df['xyz_totalshiftratio'] = df['xshift'].abs() / (df.x - df['xshift']).abs() + \
                                df['yshift'].abs() / (df.y - df['yshift']).abs() + \
                                df['zshift'].abs() / (df.z - df['zshift']).abs()            
        
    df['xyz_absoluteshift_second'] = df['xyz_absoluteshift']/df['time_localshift']
    df['xyz_totalshiftratio_second'] = df['xyz_totalshiftratio']/df['time_localshift']
        
        
        
def dfCurrent(df, period):
    #print(df.loc[ (period[0] <= df.time_local) & (df.time_local <= period[1]) ])
    df_current = df.loc[ (period[0] <= df.time_local) & (df.time_local <= period[1]) ].copy()
    
    #Append the input manual values    
    df_current['seat'] = period[4]
    df_current['experimentID'] = period[5]
    df_current.loc[df_current.dev_id == period[2], 'action'] = period[3]  
    df_current.loc[df_current.dev_id != period[2], 'action'] = 'control' 
    #important to make later calcuation in order
    df_current.sort_values(by=['time_local'], inplace=True)
    return df_current

#N is the window size
def calRollingStd(df, N):
    #calcuate rolling std
    df['x_rolling_std'] = df['x'].rolling(N).std()
    df['y_rolling_std'] = df['y'].rolling(N).std()
    df['z_rolling_std'] = df['z'].rolling(N).std()
    df['xyz_absoluteshift_rolling_std'] = df['xyz_absoluteshift'].rolling(N).std()
    
    #pick up the max records based on x, y, z and xyz
    max_x = df.loc[df['x_rolling_std'].idxmax()].to_frame().T
    max_y = df.loc[df['y_rolling_std'].idxmax()].to_frame().T
    max_z = df.loc[df['z_rolling_std'].idxmax()].to_frame().T
    max_xyz = df.loc[df['xyz_absoluteshift_rolling_std'].idxmax()].to_frame().T
    max_df = pd.concat([max_x,max_y,max_z,max_xyz])
    
    max_df['max_rolling_std'] = ['x','y','z','xyz']    
    max_df = max_df.astype({'x_rolling_std':'float64','y_rolling_std':'float64', \
                   'z_rolling_std':'float64','xyz_absoluteshift_rolling_std':'float64'})
    
    return max_df


def window_extract(df, WINSIZE, stable_condition ):
    signal_status_idx = df.index[df['binary_signal_status'] > 0].tolist()
    win_last_idx = df.index[-1]
    print("singal status index:{}".format(signal_status_idx))
    print("win_last_idx index:{}".format(win_last_idx))
    
    change_start_index = signal_status_idx[0]
    change_end_index = signal_status_idx[-1]
    
    window = pd.DataFrame()
    print("start index {}; end index {}".format(change_start_index, change_end_index))
    #return last stable window before signal move
    if stable_condition==1:
        window = df.loc[:change_start_index].tail(WINSIZE+1) 
        window.drop(window.tail(1).index, inplace=True)
    #return first stable window after signal move; exp 21 seems not have a clear stable window here
    elif stable_condition==2:
        if change_end_index == win_last_idx:
            window = None
        else:
            window = df.loc[:change_end_index].tail(WINSIZE+1) 
            window.drop(window.head(1).index, inplace=True)
    return window


#prepare windows features, single, pair and triple
def window_feature(win, NEGSIGNALADJUST):
    #remove minus by add x,y an z with 1000
    win[['x','y','z']]=win[['x','y','z']]+NEGSIGNALADJUST
    win_feature = win.mean()[['x','y','z']]
    win_feature['xy'] =  math.log2(win_feature['x'] * win_feature['y'])
    win_feature['xz'] =  math.log2(win_feature['x'] * win_feature['z'])
    win_feature['yz'] =  math.log2(win_feature['y'] * win_feature['z'])
    win_feature['xyz'] =  math.log2(win_feature['x'] * win_feature['y'] * win_feature['z']) 
    #print("*****inside window feature function, print winfeature:")
    #print(win_feature)
    #print("*****inside window feature function, print winfeature completed:")
    return win_feature


def win_feature_change(win_start_feature, win_end_feature):   
    
    
    win_change = win_end_feature.subtract(win_start_feature, fill_value=0)
    win_change_ratio = win_change / win_start_feature
    win_change_ratio.index += '_ratio'

    ##########
    #also include the x start, and x end in absolute value (x_start, y_start, z_start, x_end, y_end, z_end)
    ##########
    #win_start[]
    
    win_change_combined = win_change.append(win_change_ratio)
    #print(win_change)
    #print(win_change_ratio)
    return win_change_combined
```

```python
import math
#find the max std for x, y and z with the give window size (5)

# (df_current_target.loc[df_current_target['xyz_absoluteshift_second'].idxmax()].to_frame().T)


#print(df_agg_rolling_std.dtypes)
#display(df_agg_rolling_std.loc[df_agg_rolling_std.max_rolling_std =='xyz'])
largest_row = df_agg_rolling_std.loc[df_agg_rolling_std['xyz_absoluteshift_rolling_std'].idxmax()].to_frame().T
largest_std = float(largest_row['xyz_absoluteshift_rolling_std'])*(1+NOISEBUFFER)
print("after adjustment scalue up {}, largest std is:{}".format(NOISEBUFFER,largest_std))

df_train_all = pd.DataFrame()


#Annotate area of high standard dev and low standard dev using the curoff above for target
for idx, period in enumerate(datetime_objlist):
    print(idx)
    print(period[-4])
    df_current = dfCurrent(df, period)
    df_current_target = df_current.loc[df_current.action!='control'].copy()   

    calShift(df_current_target)
    #calculate the max rolling std given window size 5 from controlgroup.
    #assumption is only std larger than this, is potentially the car move in/out
    max_df = calRollingStd(df_current_target,WINSIZE)
    #compare the xyz_absoluteshift_rolling_std with the cutoff value 
    df_current_target.loc[df_current_target['xyz_absoluteshift_rolling_std'] > largest_std,'binary_signal_status'] = 100 
    df_current_target['binary_signal_status'] = df_current_target['binary_signal_status'].fillna(0)

    #display(df_current_target);add extraline (based on stdv learned from control group)
    #please note, for more robust, algorithm can be adjusted "binary_signal_status", to filter out single node of high std.
    tz = timezone('Australia/Sydney') 
    fig, axs1 = plt.subplots(1, 1, figsize=(18,5))
    formatFig(axs1, df_current_target, 1)    

    #now extract signal windows (before and after); 
    #1: get the last stable point before fluctuate; 2: get the first stable point after fluctuate
    win_start = window_extract(df_current_target, WINSIZE, 1)
    win_end = window_extract(df_current_target, WINSIZE, 2)
    
    #if both start and end exists ....
    if (win_start is None) or (win_end is None):
        print("There is null window at index {}".format(idx))
        continue
        
    
    #for x, y and z axis, calculate the median value
    win_start_feature = window_feature(win_start,NEGSIGNALADJUST)
    win_end_feature = window_feature(win_end,NEGSIGNALADJUST)

    #GP1, 2 features; not consider GP3 changing from x to y (assuming that's not relevant)
    win_change = win_feature_change(win_start_feature, win_end_feature)
    win_change['class'] = period[-4]
    
    #pivot and prepare as input record for training (separating park and move)
    #display(win_change.to_frame().T)
    df_train_all = df_train_all.append(win_change.to_frame().T)
    

    #display(df_current_target)
    #display(df_current_control)            
    #display(df_agg_rolling_std)
    #display(signal_start)
    #display(signal_end)    
    
    #break

```
